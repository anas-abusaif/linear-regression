```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
students=pd.read_csv('./students-performance.csv')
students.head()
```

```python
students.info()
```

```python
students.describe()
```

```python
plt.scatter(students['reading score'],students['writing score'])
```

```python
x=students['reading score'].values
y=students['writing score'].values
```

```python
x=x.reshape(-1,1)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.5, test_size=0.5, random_state=100)
lr = LinearRegression()
lr.fit(x_train,y_train)
y_predict = lr.predict(x_test)
```

```python
print(f"Train accuracy {round(lr.score(x_train, y_train) * 100, 2)}%")
print(f"Test accuracy {round(lr.score(x_test, y_predict) * 100, 2)}%")

plt.scatter(x_train,y_train, color='pink')
plt.plot(x_test,y_predict)
plt.xlabel('reading score')
plt.ylabel('writing score')
plt.title("Trained Model")
```
